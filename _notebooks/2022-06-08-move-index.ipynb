# Analysis of MOVE Index versus Bond Prices

# Overview

As with many Hedgeye subscribers in the spring of 2022 I have a portion of my portfolio invested in government bond ETFs. Also along with other Hedgeye subscribers I look at the ^MOVE index to see what direction bonds are likely to move in the future. How well does the ^MOVE index correlate to bond prices? Does a significant move in the ^MOVE index mean bonds will increase in price?

# TL;DR

Yes, the direction of the ^MOVE index can be used as an indicator of bond price movement, and a more extreme move in the MOVE index in the negative direction correlates more strongly to a positive move in bond prices.

```python
import pandas as pd
```

To get data for my analysis I visited finance.yahoo.com and downloaded historical data for the ^MOVE index, IEF (7-10 year treasury bond ETF) and TLT (20+ year treasury bond ETF). This data could have all been combined into one dataframe using code, but for the sake of laziness, I simply copied and pasted the data for the closing price for each asset going back as far as data would go (Nov 13th 2002). 

```python
data_df = pd.read_csv('move_tlt_ief.csv')
```

```python
data_df.head()
```

My first and simpliest analysis was to look at a simple correlation. As expected, the MOVE index and bond prices are inversely correlated. The MOVE index is a measure of how risky bond investors precieve investing in bonds to be. We would expect that as bond investing gets less risky, that bond prices would go up, which is exactly what we find to be the case here. 

```python
data_df.corr(method='pearson')
```

Next I created a new column in order to track the rate of change of the MOVE index over time. The value in each individual row of the dataframe is the percentage change in the MOVE index over the previous 20 datapoints. I chose 20 days, which is pretty arbitrary, but it seems like the right number to use right? If anyone has a smarter way to choose this number, please hit me up [@paulknick](https://twitter.com/Paulknick) on Twitter. 

```python
data_df["change"] = data_df.MOVE.pct_change(periods=20) # 20 days
```

```python
data_df.tail(20)
```

Next I wanted to look at the most extreme examples, so I started with the 25% most extreme moves in the MOVE and found that it roughly equaled -10%.

```python
data_df.describe()
```

I made a dataframe that just contained the samples that had only changed greater than -10% over 20 days. That dataframe was called "extreme_df".

```python
extreme_df = data_df[(data_df.change < -0.1)]
```

The correlation for this more extreme dataset was -.53. Remember that for our base dataset the correlation was -.43. This outcome was also expected, the more extreme the change in the MOVE in the negative direction resulted in more extreme moves in the price of bond ETFs in the positive direction. 

```python
extreme_df.corr(method='pearson')
```

Next I did the same thing again, getting the top 25% of the 25% (so 6.25%) most extreme moves in the MOVE in the negative direction. This roughly equated to a 20% move in the MOVE index over a 20 day period. 

```python
extreme_df.describe()
```

```python
extreme_extreme_df = extreme_df[(extreme_df.change < -0.2)]
```

```python
extreme_extreme_df.describe()
```

Here we see the same outcome as before, but less of a change. As we went from 100% of the sample size, down to 25% of the sample size, then further down to 6.25% of the sample size we have seen the correlation move -.43 &rarr -.53 &rarr -.59 for IEF and -.44 &rarr -.53 &rarr -.54 for TLT. 

```python
extreme_extreme_df.corr(method='pearson')
```
